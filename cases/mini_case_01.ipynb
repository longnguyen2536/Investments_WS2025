## Mini Case 01: The Phoenix Foundation – Asset Allocation in Crisis

### The Scenario
Date: January 15, 2023

The Phoenix Foundation is a mid-sized university endowment managing $500 million (currently marked down to $425 million after a brutal 2022). For twenty years, the fund relied on a rigid Strategic Asset Allocation (SAA) of 60% Equities and 40% Long-Term Bonds. This "balanced" approach was built on the assumption that bonds would hedge equity risk.

However, in 2022, inflation spiked, the Federal Reserve hiked rates, and the correlation between stocks and bonds flipped from negative to positive. Both asset classes crashed simultaneously.

Normally, the Foundation covers its annual scholarship budget through a mix of investment income and new alumni donations. However, donations have dried up. The University President has issued a non-negotiable directive: *"We have a hard liability of $25 million due in exactly one year for student scholarships. This must be paid from the endowment principal. We cannot miss this payment."*

While the endowment’s mission is perpetual (infinite horizon), this immediate cash need has created a severe duration mismatch against their long-term bond holdings.

### The Committee (Group Roles)
*Your group represents the Investment Committee. The Board has a Risk Aversion score of A = 3.0.*

1. The Chair (The Traditionalist)
*   Philosophy: "Stay the Course."
*   Perspective: You view the Strategic Asset Allocation (SAA) of 60/40 as a binding constitution. You argue that the endowment is a perpetuity; selling assets at the bottom to satisfy a short-term need is a disastrous market-timing mistake. You define the Investment Universe strictly as high-quality US Equities and Treasuries.
*   Goal: Prevent the committee from reacting emotionally. Maintain the long-term strategy.

2. The CIO (The Modernist)
*   Philosophy: "The Frontier Has Moved."
*   Perspective: You argue that the SAA failed because the fundamental correlation structure of the market has changed. You believe the "Safe Bucket" (Bonds) is no longer safe because it is now correlated with the "Growth Bucket" (Stocks). You want to re-optimize the portfolio weights immediately based on the new math.
*   Goal: Prove that the current portfolio is inefficient and restructure the risk assumptions.

3. The Treasurer (The Pragmatist)
*   Philosophy: "Cash is King."
*   Perspective: You are focused entirely on the $25 million liability. You view the Long-Term Bonds not as an asset, but as a massive risk. You propose a Tactical Asset Allocation (TAA) shift: set aside enough capital *today* into 1-year T-Bills to fully guarantee the payment next year.
*   Goal: Immunize the Year 1 liability immediately to ensure solvency, regardless of the impact on long-term growth.

4. The Active Member (The Manager Selector)
*   Philosophy: "Alpha Solves Everything."
*   Perspective: You championed Apex Capital, the star stock-picking firm that runs the equity portfolio. Apex is down, but you insist they are "high conviction" investors. You believe the committee is focusing too much on Beta (market movement) and ignoring the value of skill (Alpha).
*   Goal: Defend Apex Capital. Prove via CAPM that their performance is due to skill, not just high risk, and prevent them from being fired to cut costs.

### Analysis & Discussion Points

#### 1. Strategic Allocation & Utility (The Foundation)

a. The Baseline: Calculate the Expected Return and Variance of the current 60/40 Stock/Bond portfolio using the "Current Regime" correlation (+0.50). *(Note: Use Apex Net Return for Stocks and $E(r_{Bond})$ for Bonds).*

b. The Scorecard: Calculate the Utility Score ($U = E(r) - 0.5 \times A \times \sigma^2$) for this current portfolio using the Board's risk aversion of $A=3$.

c. The Treasurer's Challenge: The Treasurer proposes determining the Present Value of the $25M liability ($r_f=4\%$), moving that exact amount into T-Bills, and holding the remainder in the 60/40 mix. Does this new mix (approx. 6% Cash / 56% Equity / 38% Bond) increase or decrease the Board's Utility?

#### 2. The Fixed Income Dilemma (Immunization)

a. The Price Shock: If interest rates rise another 100 basis points (1%) tomorrow, how much more value will the bond portfolio lose? Use Duration and Convexity to calculate the estimated percentage drop.

b. The Trade-off: The bond portfolio has a duration of 16.5 years.
    *   Explain the Price Risk facing the Year 1 scholarship payment if we hold the long bonds.
    *   Explain the Reinvestment Risk facing the endowment's long-term perpetuity (Year 20+) if the Treasurer switches everything to T-Bills.

#### 3. Expanding the Universe (The CIO's Proposal)

a. The Hurdle: The CIO wants to add the "Real Assets Fund" (Exhibit D). Compare the Sharpe Ratio of this new fund to the Sharpe Ratio of the current Market Index.

b. The Diversification Effect: Even if the Real Assets Fund has a lower return than Stocks, how does its low correlation (0.10) justify adding it to the portfolio? (Explain qualitatively or using the Portfolio Variance formula).

#### 4. The Active Management Verdict

a. Net Returns: Calculate Apex's Expected Return Net of Fees.

b. The CAPM Test: Calculate the Required Rate of Return for Apex Capital using the CAPM.

c. The Alpha: Calculate Jensen’s Alpha for Apex using the Net Return. Is the Active Member right (skill) or is the Treasurer right (just high beta)?

d. The Optimization (The Information Ratio):
    *   *Technical Guidance:* To maximize the Sharpe Ratio when combining an active fund with a passive index, we assume the active fund acts as a "beta-neutral overlay" to the market. The efficiency gain depends on the Information Ratio (IR): $IR = \frac{\alpha}{\sigma(e)}$.
    *   The Formula: $S_{New}^2 = S_{Market}^2 + IR_{Apex}^2$.
    *   Task: Calculate the Information Ratio for Apex. Use the formula above to prove whether adding Apex to the portfolio mathematically increases efficiency compared to holding the Market Index alone.

### The Final Verdict (Group Presentation)

Your group must present a unified recommendation to the University President. Choose ONE strategic path.

*   Path A (Defensive / Tactical): Side with the Treasurer. Liquidate enough long bonds to buy T-Bills equal to the PV of the liability. Fire Apex (move to Index) to reduce fees and volatility.
*   Path B (Strategic / Traditional): Side with the Chair. Maintain the 60/40 SAA. Keep the long bonds (betting on mean reversion). Keep Apex (betting on Alpha).
*   Path C (Modern / Diversified): Side with the CIO. Add the "Real Assets Fund" to the mix to fix the correlation problem. Re-optimize weights to maximize Utility.
